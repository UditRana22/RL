```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x= np.random.randint(0,50,1000)
y= x + np.random.normal(0,10,1000)
```

```python
np.corrcoef(x,y)
```

```python
plt.scatter (x,y)
plt.show
```

```python
y= 100-(x + np.random.normal(0,10,1000))
```

```python
np.corrcoef(x,y)
```

```python
plt.scatter (x,y)
plt.show()
```

```python
df = pd.DataFrame({'a' : np.random.randint(0,50,1000)})
df['b'] = df['a']+np.random.normal(0,10,1000)
df['c'] = 100 - df['a'] + np.random.normal(0,50,1000)
df['d'] = np.random.randint(0,50,1000)
df.corr()
```

```python
pd.plotting.scatter_matrix(df,figsize=[6,6])
plt.show()
```

```python
from pandas.io.feather_format import RangeIndex
plt.matshow(df.corr())

plt.xticks(range(len(df.columns)), df.columns)
plt.yticks(range(len(df.columns)), df.columns)
plt.colorbar()
plt.show()
```
